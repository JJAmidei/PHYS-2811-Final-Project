# Length Contraction

One of the stranger effects in SR, that follows as a natural consequence of time dilation is length contraction. Since the laws of physics are valid in all inertial reference frames, we can define length as $\mathbf{length} \hspace{0.1cm} = \hspace{0.1cm} \mathbf{velocity} \hspace{0.1cm} \times \hspace{0.1cm} \mathbf{time}$. [1]

Once again, we imagine a train that is travelling at some velocity $\vec v$ East, while I am a passenger and you stand off the side watching it pass. Your frame on the ground is $S$ and my frame (the frame of the moving train) is $S'$. 

![traingroundframe](https://imgur.com/ohSnLmI.gif)

For me, measuring the length of the train is easy since I'm in its rest frame. I can take as much time as I’d like measuring it and making sure I get a precise length. However, since it's moving in your frame and you can’t take a ruler to the side of the train, we devise a method of measuring its length that takes advantage of the relationship between length, velocity, and time. We already know the velocity, so we decide to see how long it takes for the full length of the train to pass a particular point. We pick some point $P$ on the tracks, wait for the front of the train to pass it when we immediately start our timers, and then wait for the back of the train to pass the same point when we immediately stop our timers. 

![traingroundframewline](https://imgur.com/rsBzHEN.gif)

After we both measure our times, we come together and compare. Immediately, we see that our times are different. Since my frame $S'$ was moving relative to point $P$ on the tracks, I end up measuring a greater interval of time has passed than you did in frame $S$. This makes sense, since both events that we are comparing (the front of the train passes $P$ then the back of the train passes $P$) happened at the same spot in your frame, meaning that your frame $S$ is the rest frame and your time is the proper time. We use our times to calculate each length and we end up getting this. 

$$ (S): \hspace{0.1cm} l = v \Delta t \hspace{2cm} (S'): \hspace{0.1cm} l' = v t $$

$$ (S): \hspace{0.1cm} l = v \Delta t \hspace{1.5cm} (S'): \hspace{0.1cm} l' = v \frac{\Delta t}{\sqrt{1 - \frac{v^2}{c^2}}} $$

$$ l' \sqrt{1 - \frac{v^2}{c^2}} = v \Delta t $$

$$ l' \sqrt{1 - \frac{v^2}{c^2}} = l $$

From this we can see that the length is changed from frame to frame by the same factor as time, but instead of being dilated, its position in the numerator tells that the magnitude of the length will become smaller as the relative speed increases (i.e. the length will be contracted). One important aspect is that the rest frame of the object that’s length is being measured gives us what is called the proper length. In the above scenario, the train was being measured, so the length in the rest frame of the train (i.e. my frame) gave us the proper length. In your frame, even though you had the proper time, the events were in your frame, so the train’s rest frame was moving relative to you so you experienced a contracted length. [2]

We can see the relationship between the velocity and the length in the plot in the cell below.

```python
import matplotlib.pyplot as plt
import numpy as np

c = 3e8

# Similar to the time plot, you can change the values for the x-axis by changing the first 2 terms
# np.arange(lowest x value, highest x value, step size)
# The line directly below will allow you to look at the just the top 90% of velocities
# just make sure to comment out the line below it before running the code again
# v_array = np.arange(0.9*c, c, 1000) / c
v_array = np.arange(0, c, 1000) / c

inv_gamma_array = np.sqrt(1 - v_array ** 2)

plt.plot(v_array*c, inv_gamma_array)
plt.xlabel('Velocity (m/s)')
plt.ylabel('Effect of the gamma factor/length contraction factor')
plt.title('Effect of the gamma factor/length contraction factor vs. Velocity')
plt.show()
```

He we can see the relativistic effects start becoming more pronounced at lower velocities than with time dilation; however they have to be quite large. Based on this plot, length contraction seem to start becoming substantial around about one-sixth of the speed of light, contracting at an increasing rate from then on till it reaches zero as it approaches the speed of light. 

When we consider what would happens to length and distance for an object that's approaching the speed of light, this would seem to tell us that the distance between events and points would shrink more and more till the point where they could be substantially smaller than they would be in their rest frame.   

As another question to the reader, what would the contraction of length mean for travel at the speed of light? Would distance still be a meaningful term?


In the cell below there's a calculator you can use to see what the contracted length and influence of the gamma factor would be given a velocity and proper length. As with the time dilation calculator, this calculator cannot handle any speed equal to or greater than $299,792,458$, but I encourage you to see what happens to the proper length as the velocity approaches the speed of light.

```python
import numpy as np

c = 3e8

v = float(input("Enter the relative velocity of the moving frame (meters per second): "))
l_p = float(input("Enter the length the object in its rest frame (meters): "))

gamma = 1 / np.sqrt(1 - ((v**2)/c**2))
l = l_p / gamma

print("The length contraction factor is:", 1 / gamma)
print("The length in the moving frame is:", l, "m.")
```
