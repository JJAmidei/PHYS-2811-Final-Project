# Measurement of Time 

It was taken for granted in classical physics that there was a universal time. Afterall, this seems to be a reasonable belief from our direct experience. If I left my house and travelled twenty minutes to come meet you somewhere while you were sitting down and waiting, we would both experience the same twenty minutes. Despite being in different locations and having different inertial frames, where mine moving relative to yours, there seems to be a shared time. There is simply a natural and universal rate at which time progresses and events happen in a particular sequence in that time. 

As it turns out, this is not correct. As we will see, following from the postulates of special relativity, travel at close to the speed of light has the potential to drastically change real physical measurements, like time, in a way Newtonian mechanics could not have predicted.


# Time Dilation

Once again, imagine a train car. Except this time, on the floor of the train car there is a lightbulb that emits a flash of light straight up towards the ceiling where it is reflected back by a mirror, some height $h$ above.

![trainwlight](https://imgur.com/e1Ub9pn.gif)

After one flash, the emitted light travels a vertical distance $h$ up to the mirror, and then back down $h$ in some interval of time $\Delta t$. Since the speed of light is constant, we can solve for the time it will take for the light to travel the full distance of $2h$.

$$ c = \frac{2h}{\Delta t} $$

$$ \Delta t = \frac{2h}{c} $$

Specifically, this is the time in the train’s rest frame. Now let's see how long this will take in the frame of an outside observer; namely, you. 

You are standing right next to the train, watching it pass by at some velocity $\vec v$ to the East, while the light bulb flashes on the inside. 

![groundframetrainwlight](https://imgur.com/ahH5SmB.gif)

Since the train is moving, the path that it travels will be a bit different. 

![groundframepathoflight](https://imgur.com/gIrdgBB.png)

Here we can see that the distance along the track, $x$, can be defined as the velocity of the train times the interval of time it takes for the light to travel from the light bulb to the mirror ($ x = \vec v \frac{t}{2}$). We can also see that the distance that the light travels, $r$, will be equal to the speed of light times the same time interval ($r = c \frac{t}{2}$). We already know that the vertical distance is $h$, so we can use vector addition to get the follow equation. 

$$ (\frac{ct}{2})^2 = h^2 + (\frac{vt}{2})^2 $$

Then solving for $t$

$$ (1) \hspace{0.1cm} h^2 = (\frac{ct}{2})^2 - (\frac{vt}{2})^2 \hspace{1.5cm} (2) \hspace{0.1cm} h = \sqrt{(\frac{ct}{2})^2 - (\frac{vt}{2})^2} \hspace{1.5cm} (3) \hspace{0.1cm} h = \frac{t}{2} \sqrt{c^2 - v^2} $$

$$ (4) \hspace{0.1cm} t = \frac{2h}{\sqrt{c^2 - v^2}} \hspace{1.5cm} (5) \hspace{0.1cm} t = \frac{2h}{c\sqrt{1 - \frac{v^2}{c^2}}} $$

Remember that the time in our train's frame was $\Delta t = \frac{2h}{c} $. So, plugging that in, we get

$$ t = \frac{\Delta t}{\sqrt{1 - \frac{v^2}{c^2}}} $$

[1]

What does this equation tell us? First, we should look at the denominator. We have $\sqrt{1 - \frac{v^2}{c^2}}$, where $v$ is the speed of the frame of the train car relative to us and $c$ is the constant speed of light in a vacuum. As the speed of the train’s frame relative to us approaches the speed of light, the square root will approach $0$. According to the equation we just derived, this will result in time in our frame getting bigger and bigger as it approaches infinity. In other words, as the velocity of the moving frame increases, so does the interval of time between events in our frame. This effect is called time dilation. 

The term $ \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}} $ is commonly called the Lorentz or gamma factor and is represented by the lower case Greek letter $\gamma$. Below, you can see a plot that shows the relationship between the velocity of the moving frame and the Lorentz/gamma factor. 


```python
import matplotlib.pyplot as plt
import numpy as np

c = 3e8

# You can change the values in for the x-axis by changing the first 2 here
# np.arange(lowest x value, highest x values, step size)
# for example, delete the # at the begining of the next line to see the top 90% of speeds
# just make sure to commect out the line below it 
# v_array = np.arange(0.9*c, c, 1000) / c 
v_array = np.arange(0, c, 1000) / c  

gamma_array = 1 / np.sqrt(1 - v_array**2)

plt.plot(v_array * c, gamma_array)
plt.xlabel('Velocity (m/s)')
plt.ylabel('effect of the gamma factor/time dilation factor')
plt.title('Effect of gamma factor/time dilation factor vs Velocity')
plt.show()
```

As we can see, the gamma factor is barely non-zero for most speeds. Even at mind bendingly large velocities the relativistic effects may be minimal. However, as we start to approach the upper end of the plot (somewhere around 90% of $c$), the magnitude of the gamma factor begins to increase rapidly and the relativistic effects begin to become very pronounced. 

Interestingly, this plot shows us that there is an asymptote at the speed of light. This leads to one of the most profound revelations from special relativity. That is that the speed of light cannot be exceeded, only approached. Therefore, the speed of light is not just unique because it is constant in all reference frames, but also because it serves as a universal speed limit (i.e. a maximum propagation speed).  

What does the plot tell us when we consider what happens to time for an object approaching the speed of light? When we think of the example of the light on the train, we see that the time between two events in which took place in a non-moving frame was dilated for all other frames which had motion relative to it. As we can see, this effect would be massively amplified as a frame’s speed approaches the speed of light. Additionally, we can remember that it would be valid to treat the train as stationary, with the ground frame moving relative to it. This would be true of any object travelling near the speed of light. Meaning that, if an object is travelling at near the speed of light in an initial frame, we could treat it as the rest frame, where the initial frame is travelling relative to it at near the speed of light. So while the time that the object experiences is massively dilated in the initial frame, the time in the initial frame is also dilated for the object.
As a question to the reader, what would this imply about travel equal to the speed of light if an object was able to accelerate up to this point?


Below is a calculator that will return the values of the gamma factor and dilated time given a relative velocity and a time interval in the moving frame. Try to plug in a large value of $v$ to see how the time changes. Just make sure the velocity you enter does not exceed $299,792,458$, otherwise the calculator will break. 

```python
import numpy as np

c = 3e8

v = float(input("Enter the relative velocity of the moving frame (meters per second): "))
t_p = float(input("Enter the time in the moving frame (seconds): "))

gamma = 1 / np.sqrt(1 - ((v ** 2)/c**2))
t = gamma * t_p

print("The magnitude of the gamma factor is:", gamma)
print("The dilated time is:", t, "s.")
```


On a final note, in the equation we derived for dilated time above, the term $\Delta t$ represents what’s called the proper time. Specifically, the proper time is the shortest possible interval of time between two events. The time between the light bulb emitting a flash and the mirror reflecting the flash cannot be shorter than it is in the train's frame, so it's the proper time. Another way of looking at this is that the interval of time is defined by the time between two events that happened in the same place (i.e. the flash and the light from the flash hitting the bulb again after being reflected by the mirror). These events are connected by location because they happened in the rest frame. On the other hand, in the frame of an outside observer moving relative to the train, these events happened at three different places and there was a time dilation. Therefore, we could define the proper time as the time in the rest frame (i.e. the frame where two or more events happen at the same place.)
