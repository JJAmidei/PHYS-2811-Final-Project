# Momentum

When we consider classical Newtonian momentum, it is fairly well defined. According to Newton's 2nd law, we get $\vec F = m\vec a$, where $\vec a$ is equal to the change of velocity over time. This allows us to rewrite the equation as 

$$ \vec F = m \vec a = m \frac{d \vec v}{dt} $$

We can solve for momentum by integrating both sides. 

$$ \int \vec F dt = m \int \frac{d \vec v}{dt} dt $$

$$ \vec p = m \vec v $$

This shows us that momentum and velocity have a linear relationship with one another. Classically, there was no absolute upper limit to velocity, so this linear relationship was thought to be valid in all cases. However, as it follow from Maxwell's equations, this is not the case. With the speed of light serving as a universal speed limit , we get the relationship we see on the plot in the cell below.  


```python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8
m = 1 
v = np.linspace(0, 4*c, 1000)

p_lin = m*v
p_rel = np.zeros_like(v)
for i in range(len(v)):
    if v[i] < c:
        p_rel[i] = m * (v[i] / np.sqrt(1 - (v[i]**2 / c**2)))
    else:
        p_rel[i] = np.nan  

plt.plot(p_lin, v, 'r', label='Classical Momentum')
plt.plot(p_rel, v, 'b', label='Relativistic Momentum')
plt.axhline(y = c, linestyle='--', color='grey', label='Speed of Light')
plt.legend()

# you can zoom in by changing the limits of each axis by 
# changing the values in both plt.xlim and plt.ylim if you so wish
plt.xlim(0, 4*c)
plt.ylim(0, 4*c)
plt.xlabel('Momentum')
plt.ylabel('Velocity')
plt.title('Momentum vs Velocity')
plt.show()
```

How can we find this mathematically? First, we go back to the first postulate of relativity: the laws of physics hold in all reference frames. From here, we can assume that if momentum is conserved in one frame, it will be conserved in all other frames as well. 

Taking the definition of momentum in the rest frame where $t_{0}$ is the proper time, we get

$$ \vec p = m \frac{d\vec v}{dt_{0}} $$ 

Here we can compare the value of the proper time to the time in some primed frame. Defining $ \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}} $ as $\gamma$, we get

$$ \gamma dt_{0} = dt $$

Dividing over $\gamma$, we are to compare the momentum in our rest frame to the momentum in the primed frame. 

$$ dt_{0} = \frac{dt}{\gamma} \hspace{0.2cm} \rightarrow \hspace{0.2cm} m \frac{d\vec v}{dt_{0}} = \gamma m \frac{d\vec v}{dt}$$

Which gives us the relationship we see in the plot above.
 


# Energy 

From the momentum, we can find the energy easily enough. Since energy is $ E = \int \vec F \cdot d\vec r$ and the $ \vec F = \frac{d\vec p}{dt}$, we can establish a connection between momentum and energy. Remembering that our relativistic momentum was defined as $\gamma m \frac{d\vec v}{dt}$, we get  

$$ E = \int \frac{d \vec p}{dt} d\vec r =  \int \frac{dr}{dt} d\vec p = \vec v \vec p = \gamma m \vec v^2$$

As it turns out, this is close to the proper expression. [1] In order to make this completely correct, we have to change the velocity to $c$, giving us 

$$ E = \gamma m c^2 = \frac{m c^2}{\sqrt{1 - \frac{v^2}{c^2}}}$$

Then we can use binomial approximation to show that 

$$ E = \frac{m c^2}{\sqrt{1 - \frac{v^2}{c^2}}} \approx mc^2(1 + \frac{1}{2} \frac{v^2}{c^2}) = mc^2 + \frac{1}{2}mv^2$$

This tells us that for slowly moving masses (less than roughly 10% the speed of light), we can approximate an objects energy as some constant term $mc^2$ and the classical kinetic energy, $\frac{1}{2}mv^2$. 

In both the original equation and the approximation, we can see that when there is no motion, we are left with the term $mc^2$. This is called the rest energy, which is a very important part of relativistic energy, albeit not extremely relevant for our purposes now.

When a body is not at rest, we can think of its total energy $\gamma m c^2$ as the rest energy plus some kinetic energy. This gives us 

$$ E = mc^2 + K$$

Rearranging, we get

$$ K = E - mc^2 = \gamma mc^2 - mc^2 = mc^2(\gamma -1) $$

which gives us the relativistic kinetic energy in terms of the rest mass and the gamma factor. 

Before moving on, we can obtain one more relation by comparing the energy and momentum again. We start with

$$ E = \frac{m c^2}{\sqrt{1 - \frac{v^2}{c^2}}} \hspace{0.3cm}  \hspace{0.3cm} \vec p = \gamma m \frac{d\vec v}{dt}$$

Then we divide the energy from the momentum and get

$$ \frac{\vec p}{E} = \frac{\vec v}{c^2}$$

and 

$$ \frac{v}{c} = \frac{\vec pc}{E} $$. 

Finally, when we square both sides we get

$$ E^2 = (pc)^2 + (mc^2)^2 $$

This shows us how momentum, rest mass, and total energy relate to one another. This relationship is sometimes called the *Pythagorean relation* since each component fits together mathematically like the sides of a right triangle. [1]

# Atmospheric Muons 

So why did we take the time to work through everything that we did in the last section? Because in order to talk about this specific example, we need to be acquainted with time dilation, length contraction, relativistic momentum, and energy. 

A muon is a negatively charged elementary particle, with a mass of $105.66 \hspace{0.1cm} MeV/c^2$. Importantly, muons are unstable, only having a mean half-life of $2.2\mu s$. Additonally, roughly 10,000 muons reach every square meter of the Earth's surface each minute. This is caused by the breakdown of pions produced by cosmic rays that hit the upper atmosphere; on average 10-15 km above the surface of the Earth.  [3]

![cosmicray](https://imgur.com/dSr9NK1.png)

The mean energy of a muon at sea level is about $4 \hspace{0.1cm} GeV$, losing roughly $2 \hspace{0.1cm} GeV$ of energy by passing through the atmosphere. [3]

Using the equation for relativistic kinetic energy, we can find the velocity.

$$ K = mc^2(\gamma - 1)$$

$$ \frac{K}{mc^2} = \gamma - 1 \rightarrow \frac{K}{mc^2} + 1 = \gamma $$

Setting $K$ equal to $4 \hspace{0.1cm} GeV$, we can solve for the gamma factor. 

$$ \frac{4 \hspace{0.1cm} GeV}{105.66 \hspace{0.1cm} MeV} + 1 = \gamma \approx 38.74$$

Then solving for the velocity from the gamma factor, we get

$$ 38.74 = \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}} \rightarrow \frac{1}{38.74} = \sqrt{1 - \frac{v^2}{c^2}} $$

Since the whole equation is now known constants, we can calculate the velocity which is $v = 0.9997c$, or roughly 99.97% the speed of light. 

$$ \frac{1}{(38.74)^2} = 1 - \frac{v^2}{c^2}  \rightarrow \frac{v^2}{c^2} = 1 - \frac{1}{(38.74)^2}$$ 

$$ v^2 = (1 - \frac{1}{(38.74)^2})c^2$$ 

$$ v = \sqrt{1 - \frac{1}{(38.74)^2}}c$$ 

which gives us $v = 0.9997c$, or roughly 99.97% the speed of light.

Taking the muon's mean lifespan and multiplying it by the velocity we just calculated, we get that the average distance a muons travel is 

$$ 0.9997c \hspace{0.1cm} \frac{m}{s} \times 2.2 \hspace{0.1cm}\mu s \approx 660 \hspace{0.1cm}ms $$

This is obviously a far smaller distance than the 10-15 km that a muon must travel to reach the Earth's surface. This discrepancy exists because we were trying to use classical mechanics to solve a relativistic problem. When we use the equations for time dilation and length contraction, we will be able to correct for this. 

First, let's take the equation for time dilation. 

$$ t = \gamma \Delta t $$

The $\Delta t$ term refers to the proper time, which is the time in the muon's frame. This makes sense because there are two events that we are measuring the time between: (1) the muons creation in the upper atmosphere and (2) the muons detection on the surface of the Earth. In the muon's rest frame, both events could be understood as having happened at the same location, thus making the time between them in the muon's frame the proper time. 

Taking our gamma factor, which allows us to compare the muon's frame with the Earth's frame, we just have to multiple the two values by each other to get the dilated time. 

$$ \gamma \Delta t = 38.74 \times 2.2 \hspace{0.1cm} \mu s \approx 85.2 \hspace{0.1cm} \mu s$$

When we take this dilated time and multiple it by the muon's velocity we get 

$$ 0.9997c \hspace{0.1cm} \frac{m}{s} \times 85.2 \hspace{0.1cm} \mu s \approx 25,600 \hspace{0.1cm} m $$

Which more than makes up for the original discrepancy between the classical distance calculation and the muon measurements. Though the new distance is about 10 km greater than the original average distance from earlier, it is still well within the range of the atmosphere, and it is entirely possible a muon could be created that much further up. [3]

To wrap up, let's check the length contraction due to the muon's motion. Since the distance in the Earth's frame is the proper distance, we can easily make use of the equations we derived earlier.

$$ L = \frac{\Delta L}{\gamma} $$

Since we just found above that the real distance the muon travelled was 25 km, let's place this as our value of the proper length ($\Delta L$). Plugging in our values, we get 

$$ L = \frac{25.6\hspace{0.1cm}km}{38.74} \approx 660 \hspace{0.1cm}m$$

which is approximately the same value we get when we multiply the velocity by the proper time. 

What this shows is that, when the velocity of an object is large enough, classical mechanics is unable to help us describe it. Talking about things like atmospheric muons would be impossible without the concept which follow from special relativity.
